```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline
import seaborn as sns
import sklearn
```

```python
df_train = pd.read_csv('/kaggle/input/vidya-car-insurance-fraud-claim/train_qWM28Yl.csv')
df_train.head(3)
```

```python
df_train.info()
```

```python
df_train['is_claim'].value_counts()
```

```python
categorical_features = [cols for cols in df_train if df_train[cols].dtype == 'object']
categorical_features
```

```python
df_train_new1 = pd.get_dummies(data=df_train, columns=[
'area_cluster',
 'segment',
 'model',
 'fuel_type',
 'max_torque',
 'max_power',
 'engine_type',
 'is_esc',
 'is_adjustable_steering',
 'is_tpms',
 'is_parking_sensors',
 'is_parking_camera',
 'rear_brakes_type',
 'transmission_type',
 'steering_type',
 'is_front_fog_lights',
 'is_rear_window_wiper',
 'is_rear_window_washer',
 'is_rear_window_defogger',
 'is_brake_assist',
 'is_power_door_locks',
 'is_central_locking',
 'is_power_steering',
 'is_driver_seat_height_adjustable',
 'is_day_night_rear_view_mirror',
 'is_ecw',
 'is_speed_alert'
], drop_first=True)
df_train_new1.head(3)
```

```python
df_train_new1.info()
```

```python
df_train_new1.drop(columns = 'policy_id', axis = 1, inplace = True)
df_train_new1.sample(3)
```

```python
from sklearn.utils import resample
#create two different dataframe of majority and minority class 
df_majority = df_train_new1[(df_train_new1['is_claim']==0)] 
df_minority = df_train_new1[(df_train_new1['is_claim']==1)] 
# upsample minority class
df_minority_upsampled = resample(df_minority, 
                                 replace=True,    # sample with replacement
                                 n_samples= 54844, # to match majority class
                                 random_state=42)  # reproducible results
# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_minority_upsampled, df_majority])
df_upsampled.head(3)
```

```python
df_upsampled['is_claim'].value_counts()
```

```python
X = df_upsampled.drop(columns = 'is_claim', axis = 1)
y = df_upsampled['is_claim']
```

```python
# train_test_split
from sklearn.model_selection import train_test_split
(X_train, X_val, y_train, y_val) = train_test_split(X, y, test_size = 0.2)
```

```python
# importing random forest classifier from ensemble module
from sklearn.ensemble import RandomForestClassifier

# creating a RF classifier
rfc = RandomForestClassifier(n_estimators = 100) 
 
# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
rfc.fit(X_train, y_train)

# performing predictions on the validation dataset
y_pred_val = rfc.predict(X_val)
```

```python
# Constructing the confusion matrix.
from sklearn.metrics import confusion_matrix
confusion_matrix(y_val, y_pred_val)
```

```python
# To compute the F1 score, simply call the f1_score() function:
from sklearn.metrics import f1_score
f1_score(y_val, y_pred_val)
```
